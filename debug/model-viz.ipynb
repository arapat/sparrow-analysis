```python
import xgboost as xgb
```

```python
bst = xgb.Booster({'nthread': 4})
bst.load_model("./xgb-r1184-t8.bin")
```

```python
%matplotlib inline
xgb.plot_tree(bst, num_trees=20)
```

```python
with open("xgb.txt", 'w') as f:
    bst.dump_model(f)
```

```python
import numpy as np
import pandas as pd

# bst.predict()
data = np.array([[1.0] * 100, [0.0] * 100])
bst.predict(
    xgb.DMatrix(data), ntree_limit=2
)
```
